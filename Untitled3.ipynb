```python
from google.colab.patches import cv2_imshow
import cv2
import numpy as np


altura=0
anchura=0
mpix=0
alc=5
# Load the image

img = cv2.imread('/content/drive/MyDrive/ejemplo.jpg')



# Convertir a escala de grises
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

# Aplicar umbral para detectar los píxeles rojos
lower_red = np.array([0, 0, 100])
upper_red = np.array([50, 50, 255])
mask = cv2.inRange(img, lower_red, upper_red)

# Encontrar los contornos de los objetos en la imagen
contours, _ = cv2.findContours(mask, cv2.RETR_TREE, cv2.CHAIN_APPROX_SIMPLE)

# Seleccionar el contorno más grande que tenga un área roja significativa
max_contour = None
max_area = 0
for contour in contours:
    area = cv2.contourArea(contour)
    if area > max_area:
        max_area = area
        max_contour = contour

# Dibujar el contorno más grande en la imagen original
if max_contour is not None:
    x, y, w, h = cv2.boundingRect(max_contour)
    mpix=alc/w
    altura=h*mpix
    anchura=w*mpix
    cv2.rectangle(img, (x, y), (x + w, y + h), (0, 255, 0), 2)
#print("altura",altura,"anchura",anchura)

# Mostrar la imagen con el contorno más grande dibujado

#cv2_imshow(img)
# Suavizar la imagen
img_smooth = cv2.GaussianBlur(img, (5, 5), 0)

# Convertir a escala de grises
gray = cv2.cvtColor(img_smooth, cv2.COLOR_BGR2GRAY)

# Umbralización
_, threshold = cv2.threshold(gray, 127, 255, cv2.THRESH_BINARY)

# Encontrar contornos
contours, _ = cv2.findContours(threshold, cv2.RETR_TREE, cv2.CHAIN_APPROX_SIMPLE)

for cnt in contours:
    # Obtener rectángulo delimitador y dibujarlo en la imagen original
    x, y, w, h = cv2.boundingRect(cnt)
    # Calcular las dimensiones del objeto en unidades reales
    object_width = w * mpix
    object_height = h * mpix
    if(object_height >1 and object_width >1 and object_width < 60 and object_height< 60):
      cv2.rectangle(img, (x, y), (x + w, y + h), (0, 255, 0), 2)

    #c=c+1
    #print(c)
    #cv2_imshow(img)

    # Calcular el ancho y el alto en unidades reales utilizando el cuadrado de papel como referencia

      print(f'Ancho del objeto: {object_width} cm')
      print(f'Alto del objeto: {object_height} cm')

# Mostrar la imagen con los rectángulos delimitadores dibujados
cv2_imshow(img)
```

```python


# Cargar la imagen
img = cv2.imread('/content/drive/MyDrive/ejemplo.jpg')
c=0
# Suavizar la imagen
img_smooth = cv2.GaussianBlur(img, (5, 5), 0)

# Convertir a escala de grises
gray = cv2.cvtColor(img_smooth, cv2.COLOR_BGR2GRAY)

# Umbralización
_, threshold = cv2.threshold(gray, 127, 255, cv2.THRESH_BINARY)

# Encontrar contornos
contours, _ = cv2.findContours(threshold, cv2.RETR_TREE, cv2.CHAIN_APPROX_SIMPLE)

for cnt in contours:
    # Obtener rectángulo delimitador y dibujarlo en la imagen original
    x, y, w, h = cv2.boundingRect(cnt)
    cv2.rectangle(img, (x, y), (x + w, y + h), (0, 255, 0), 2)
    c=c+1
    #print(c)
    #cv2_imshow(img)

    # Calcular el ancho y el alto en unidades reales utilizando el cuadrado de papel como referencia
    # Supongamos que el cuadrado de papel tiene un ancho de 'paper_width' unidades en la vida real
    paper_width = 5.0  # Cambiar a las dimensiones reales del cuadrado de papel
    pixels_per_metric =  w / paper_width # Calcular la relación píxeles por métrica

    # Calcular las dimensiones del objeto en unidades reales
    object_width = w * pixels_per_metric
    object_height = h * pixels_per_metric

    print(f'Ancho del objeto: {object_width} cm')
    print(f'Alto del objeto: {object_height} cm')

# Mostrar la imagen con los rectángulos delimitadores dibujados
cv2_imshow(img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
